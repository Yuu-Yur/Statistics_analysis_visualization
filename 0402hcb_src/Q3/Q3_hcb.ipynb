```python
import pandas as pd
import folium
import webbrowser
```

```python
 # 1) 커피빈 매장 현황에서 주소를 추출하고 주소에서 시도, 군구를 추출한다. 0402 하청빈
cafe_data = pd.read_csv("./CoffeeBean.csv", encoding="utf-8", index_col=0, header=0, engine="python")
cafe_data['광역시도'] = cafe_data['address'].str.split().str[0]
cafe_data['행정구역'] = cafe_data['address'].str.split().str[1]
cafe_data.loc[cafe_data['광역시도']=='서울',['광역시도']]='서울시'
cafe_data.loc[cafe_data['광역시도']=='서울시',['광역시도']]='서울특별시'
cafe_data.loc[cafe_data['광역시도']=='경기',['광역시도']]='경기도'
cafe_data.loc[cafe_data['광역시도']=='충북',['광역시도']]='충청북도'
cafe_data.loc[cafe_data['광역시도']=='대전시',['광역시도']]='대전광역시'
cafe_data.loc[cafe_data['광역시도']=='경남',['광역시도']]='경상남도'
cafe_data.loc[cafe_data['광역시도']=='부산시',['광역시도']]='부산광역시'
cafe_data.loc[cafe_data['광역시도']=='인천',['광역시도']]='인천광역시'
cafe_data.loc[cafe_data['광역시도']=='강원도',['광역시도']]='강원특별자치도'
cafe_data.loc[cafe_data['행정구역']=='창원시',['광역시도']]='창원특례시'
cafe_data
```

```python
 # 2) 주소 -> 위도, 경도 부분으로 데이터를 추출한다. 0402 하청빈
cafe_data.to_csv("./CoffeeBean_sidogungu.csv", encoding="utf-8", index=False)
# 이렇게 저장한 csv 를 https://www.geoservice.co.kr/ 에서 주소->좌표.shp->좌표.csv 로 변환
cafe_coordinate_data = pd.read_csv('a.csv', encoding="utf-8", index_col=0, header=0, engine="python")
cafe_coordinate_data.reset_index(inplace=True)
cafe_coordinate_data.rename(columns={'field1':'store','field2':'address','field3':'phone','field4':'sido','field5':'gungu'},inplace=True)
cafe_coordinate_data
```

```python
 # 3) 추출한 데이터를 folium 이용해 마크업 시각화를 한다. 0402 하청빈
 # 기본은 커피빈 서면점
map_CB = folium.Map(location=[35.156233328065,129.05793488419],zoom_start=15)
for i, store in cafe_coordinate_data.iterrows():
  folium.Marker(location=[store["_Y"],store["_X"]], popup=store["store"],icon=folium.Icon(color="red", icon="star")).add_to(map_CB)

map_CB.save("coffeeBeanMarked.html")
```
